```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/retail_sales_dataset.csv")
df.head()
```

```python
display(df.info())
display(df.columns)
display(df.describe())
```

```python
print("Missing values per column:")
display(df.isnull().sum())

print("\nNumber of duplicate rows:")
display(df.duplicated().sum())
```

```python
print("Value counts for Gender:")
display(df['Gender'].value_counts())

print("\nValue counts for Product Category:")
display(df['Product Category'].value_counts())
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Gender')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Product Category')
plt.title('Distribution of Product Category')
plt.show()
```

```python
print("Summary statistics for numerical features:")
display(df[['Age', 'Quantity', 'Price per Unit', 'Total Amount']].describe())
```

```python
numerical_features = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']
df[numerical_features].hist(figsize=(10, 8), bins=20)
plt.tight_layout()
plt.show()
```

```python
numerical_features = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']
correlation_matrix = df[numerical_features].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
